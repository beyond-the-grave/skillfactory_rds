# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['24'] = 'Four By Two Productions'

long_movie_names=data.copy()
long_movie_names['title_length']=long_movie_names['original_title'].map(lambda x: len(x))
companies = long_movie_names.production_companies.str.split('|').explode().unique()
sum_gen = pd.Series(dtype='float64', index=companies)
for comp in companies:
    sum_gen[comp] = long_movie_names['title_length'][long_movie_names['production_companies'].map(lambda x: True if comp in x else False)].mean()
sum_gen.sort_values(ascending=False).head()
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
answers['25'] = 'Midnight Picture Show'

long_movie_names=data.copy()
long_movie_names['title_length']=long_movie_names['overview'].map(lambda x: len(x))
movies = long_movie_names.production_companies.str.split('|').explode().unique()
sum_gen = pd.Series(dtype='float64', index=movies)
for movie in movies:
    sum_gen[movie] = long_movie_names['title_length'][long_movie_names['production_companies'].map(lambda x: True if movie in x else False)].mean()
sum_gen.sort_values(ascending=False).head()
```

# 26. Какие фильмы входят в 1 процент лучших по рейтингу?

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] = 'Daniel Radcliffe , Rupert Grint'

cast_counter=data.copy()
cast_counter['cast_split'] = cast_counter['cast'].str.split('|')
cast_counter['cast_group']=cast_counter.cast_split.apply(lambda x: list(combinations(x, 2))).tolist()
counter_list= chain(*cast_counter['cast_group'])
Counter(counter_list).most_common()
```
